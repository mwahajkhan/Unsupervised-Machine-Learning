## **Recommendation System**

> To find the closest recommendation to a movie:

  1. Creating matrix between user_id and movie title, (each corresponding cell will represent the rating for the movie by the paticular user). There will be many null value as not all users have watched all the movies.

  2. Creating correlation matrix for every movie, with every other movie based on user ratings. Using this correlation matrix to find the top matches that relates best for a paticualr movie (with having minimun 100 ratings). The result obtained will be added to the ratings dataframe of each movie.

  3. The ones without matches will be converted to "-".

### Step 1: Matrix With User Ratings For Each Movie

```python
# Create a pivot table with 'user_id' as rows, 'title' as columns, and 'rating' as values.
moviemat = df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
moviemat.head()
```

```python
#Top rated movies with their average ratings
ratings.sort_values('num_of_ratings', ascending = False).head(10)
```

### Step 2: Correlation of One Movie with Every Other Movie Based on Number of Ratings

```python
# Iterate over each movie in the ratings dataframe.
for i in ratings.index:
    # Get the user ratings for the current movie.
    movieUserRatings = moviemat[i]

    # Calculate the correlation of the current movie with every other movie based on user ratings.
    similarToThatMovie = moviemat.corrwith(movieUserRatings)

    # Create a dataframe with the correlation results.
    corr_toMovie = pd.DataFrame(similarToThatMovie, columns=['Correlation'])

    # Remove NaN values from the correlation dataframe.
    corr_toMovie.dropna(inplace=True)

    # Join the number of ratings for each movie to the correlation dataframe.
    corr_toMovie = corr_toMovie.join(ratings['num_of_ratings'])

    # Filter the correlation dataframe to only include movies with more than 100 ratings,
    # and sort them by correlation in descending order.
    result = corr_toMovie[corr_toMovie['num_of_ratings'] > 100].sort_values('Correlation', ascending=False).head()

    # If there are at least 5 highly correlated movies, update the ratings dataframe with the recommendations.
    if result['num_of_ratings'].count() >= 5:
        print(i)
        ratings.loc[i, 'FirstMovieRecommendation'] = result.iloc[1:2].index.values[0]
        ratings.loc[i, 'SecondMovieRecommendation'] = result.iloc[2:3].index.values[0]
        ratings.loc[i, 'ThirdMovieRecommendation'] = result.iloc[3:4].index.values[0]
        ratings.loc[i, 'FourthMovieRecommendation'] = result.iloc[4:5].index.values[0]
```

```python
# Save the ratings dataframe to a CSV file.
ratings.to_csv('movie_recommendations.csv')

# Download the CSV file in Google Colab.
from google.colab import files
files.download('movie_recommendations.csv')
```
